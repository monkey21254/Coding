# MATPLOTLIB

```python
import random
import matplotlib.pyplot as plt

random.seed(1)
x = list(range(100))
y = [random.randrange(1, 5000) for i in range(100)]
scale = [random.randrange(1, 50) for i in range(100)]
li = ["Visit" + str(i) for i in range(10)]

def mapping(x):
    if x % 2 == 0:
        return 'red'
    else:
        return 'yellow'

col = list(map(mapping, x))

dict1 = {
    'x': x,
    'y': y,
    's': scale,
    'c': col,
    'alpha': 0.65
}

# plt.scatter(x, y, s = scale, c = col, alpha = 0.65)
plt.scatter(**dict1)
plt.xscale('linear')
plt.xlabel('index')
plt.ylabel('random value')
plt.xticks(list(range(0, 100, 10)), li)
plt.title('For visualization')

plt.text(1, 100, ':)')
plt.grid(True)

plt.show()
```
